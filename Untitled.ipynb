# Implementation

```python
def Primal(X, y, w, lmd=1):
    """
    Primal problem loss
    """
    n = y.shape[0]
    X_w = X.dot(w)
    ys = scipy.sparse.diags(y)
    
    mx = np.maximum(np.zeros((n, 1)), 1 - ys.dot(X.dot(w)))
    mx = np.float(np.ones((1, n)).dot(mx) / n)
    return mx + lmd / 2 * np.float(w.T.dot(w))


def Dual(X, y, alpha, lmd=1):
    """
    Dual problem loss
    """
    n = y.shape[0]
    alphas = scipy.sparse.diags(alpha[:,0])
    phi_star = alphas.dot(y).mean()
    w = 1 / (lmd * n) * alphas.dot(X).sum(axis=0).T
    return phi_star - lmd / 2 *  np.float(w.T.dot(w))

def Class(X, y, w):
    y_pred = np.sign(X.dot(w))
    return np.linalg.norm(y_pred - y, ord=1) / (2 * y.shape[0])
```

```python
def sdca(X, y, n_iter, T_0=0, lmd=0, option="average", save_loss=500, option2=None, X_test=None, y_test=None):
    """
    X : training points
    y : labels
    n_iter : number of iterations
    T_0 : the last iteration not taken into the account, used only if option!="high_prob"
    lmd: regularization parameter
    option : "average", "random" as in the article; "high_prob" returns just the last value
    save_loss : save loss every .. iteration
    option2 : if == "test" save Classification error on a test set
    X_test, y_test : testing points and labels
    """
    # initialization
    n, p = X.shape
    alpha, w = np.zeros((n,1)), np.zeros((p,1))
    indexes = np.random.randint(n, size=n_iter)
    if option!="high_prob":
        #save alphas if option is average or random
        alphas = np.zeros((n_iter - T_0, n))
        
    Primal_loss = []
    Dual_loss = []
    Class_loss = []
    
    #testing error
    Class_loss2 = []
    # iter
    for itr in range(n_iter):
        i = indexes[itr]
        
        norm_x = X[i].multiply(X[i]).sum(1)
        norm_x = np.float(norm_x)
        
        X_w = np.float(X[i].dot(w))
        delta_alpha_i = y[i] * max(0, min(1, (1 - y[i] * X_w) * lmd * n / norm_x + alpha[i, 0] * y[i])) - alpha[i,0]
        
        alpha[i, 0] += delta_alpha_i
            
        w += 1 / (lmd * n) * delta_alpha_i * X[i].T
            
        if option!="high_prob" and itr >= T_0:
            # starts saving values of alpha
            alphas[itr - T_0] = alpha.copy()
        if itr % save_loss == 0:
            #history of losses
            Primal_loss.append(Primal(X, y, w, lmd))
            Dual_loss.append(Dual(X, y, alpha, lmd))
            #Class_loss.append(Class(X, y, w))
            if option2=="test":
                Class_loss2.append(Class(X_test, y_test, w))
    if option == "average":
        alpha = np.mean(alphas, axis=0)
        w = 1 / (lmd * n) * X.T.dot(scipy.sparse.diags(alpha)).sum(axis=1)
    elif option == "random":
        i = np.random.randint(n_iter - T_0)
        alpha = alphas[i]
        w = 1 / (lmd * n) * X.T.dot(scipy.sparse.diags(alpha)).sum(axis=1)
        
    return alpha, w, Primal_loss, Dual_loss, Class_loss, Class_loss2
```

```python
def Pegasos(X, y, n_iter, lmd, save_loss=500, option2=None, X_test=None, y_test=None):
    """
    X : training points
    y : labels
    n_iter : number of iterations
    lmd: regularization parameter
    save_loss : save loss every .. iteration
    option2 : if == "test" save Classification error on a test set
    X_test, y_test : testing points and labels
    """
    # initialization
    n, p = X.shape
    indexes = np.random.randint(n, size=n_iter)
    w = np.zeros((p,1))
    Primal_loss = []
    Class_loss = []
    
    Class_loss2 = []
    # iter
    for itr in range(n_iter):
        i = indexes[itr]
        step = 1 / (lmd * (itr + 1))
        y_Xw =  y[i] * np.float(X[i].dot(w))
        w = (1 - 1/ ( itr + 1)) * w + step * ((1 - y_Xw) > 0 ) * y[i] * X[i].T
        if itr % save_loss == 0:
            Primal_loss.append(Primal(X, y, w, lmd))
            Class_loss.append(Class(X, y, w))
            if option2=="test":
                Class_loss2.append(Class(X_test, y_test, w))
    return w, Primal_loss, Class_loss, Class_loss2
```

# On vÃ©rifie les bornes

On vÃ©rifie facilement que la hinge loss a une constante de Lipschitz de 1, combinÃ©e avec avec la remarque 4 du papier pour avoir:
$$\mathbb{E}[D(\alpha^*) - D(\alpha)] \leq \frac{\epsilon}{2}$$

il nous suffit d'avoir $T_0 \geq \max(0, n \log ( 0.5 \lambda n )) + \frac{4}{\lambda \epsilon}$ et $T \geq T_0 + n + \frac{1}{\lambda \epsilon}$
